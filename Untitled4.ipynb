The differential equation I solved using normal and PINN is heat transfer differential equation.
<br>
<br>
∂u/∂t = α∂^2(u)/∂x^2.
<br>
<br>
where u(x,t) is function of temperature of x and time.
<br>
α is the thermal diffusivity constant


Using Normal Neural Networks to get the solution

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
class HeatNet(nn.Module):
    def __init__(self):
        super(HeatNet, self).__init__()
        self.fc1 = nn.Linear(2, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 1)
        self.activation = nn.Tanh()

    def forward(self, x, t):
        inputs = torch.cat([x, t], dim=1)
        x = self.activation(self.fc1(inputs))
        x = self.activation(self.fc2(x))
        u = self.fc3(x)
        return u
```

```python
# Residual of the heat equation
def heat_residual(net, x, t, alpha):
    u = net(x, t)
    u_t = torch.autograd.grad(u, t, grad_outputs=torch.ones_like(u), create_graph=True)[0]
    u_x = torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x, grad_outputs=torch.ones_like(u_x), create_graph=True)[0]
    residual = u_t - alpha * u_xx
    return residual
```

```python
# Loss function with residual
def loss_function(net, x, t, alpha):
    residual_loss = torch.mean(heat_residual(net, x, t, alpha)**2)
    return residual_loss
```

```python
# True solution of the heat equation (analytical solution)
def true_solution(x, t, alpha):
    return torch.sin(np.pi * x) * torch.exp(-np.pi**2 * alpha * t)
```

```python
alpha = 0.001
net = HeatNet()
optimizer = optim.Adam(net.parameters(), lr=1e-3)
```

```python
epochs = 5000
batch_size = 128

for epoch in range(epochs):
    optimizer.zero_grad()

    # Collocation points (sample x and t)
    x_collocation = torch.rand((batch_size, 1), requires_grad=True)
    t_collocation = torch.rand((batch_size, 1), requires_grad=True)

    # Compute loss
    loss = loss_function(net, x_collocation, t_collocation, alpha)

    # Backpropagation
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f'Epoch {epoch}: Loss = {loss.item()}')

```

```python
# Now let's compare the true and predicted solutions
x_test = torch.linspace(0, 1, 100).reshape(-1, 1)
t_test = torch.full((100, 1), 0.5)  # compare at t=0.5

# Predicted solution from the trained network
u_pred = net(x_test, t_test).detach().numpy()

# True solution
u_true = true_solution(x_test, t_test, alpha).detach().numpy()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test.numpy(), u_true, label='True Solution', linestyle='--', color='blue')
plt.plot(x_test.numpy(), u_pred, label='Predicted Solution', color='red')
plt.title('True vs Predicted Solutions for Heat Equation')
plt.xlabel('x')
plt.ylabel('u(x, t)')
plt.legend()
plt.show()
```

Using PINN to get exact solution

```python
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.fc1 = nn.Linear(2, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 1)
        self.activation = nn.Tanh()

    def forward(self, x, t):
        inputs = torch.cat([x, t], dim=1)
        x = self.activation(self.fc1(inputs))
        x = self.activation(self.fc2(x))
        u = self.fc3(x)
        return u
```

```python
# Residual of the heat equation
def heat_residual(net, x, t, alpha):
    u = net(x, t)
    u_t = torch.autograd.grad(u, t, grad_outputs=torch.ones_like(u), create_graph=True)[0]
    u_x = torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x, grad_outputs=torch.ones_like(u_x), create_graph=True)[0]
    residual = u_t - alpha * u_xx
    return residual
```

```python
# Loss function that includes the physics of the heat equation
def loss_function(net, x, t, alpha, x_b, t_b, u_b, x_0, t_0, u_0):
    # PDE loss (residual of the heat equation)
    pde_loss = torch.mean(heat_residual(net, x, t, alpha)**2)

    # Boundary loss (enforcing boundary conditions u(0, t) = u(1, t) = 0)
    u_b_pred = net(x_b, t_b)
    boundary_loss = torch.mean((u_b_pred - u_b)**2)

    # Initial condition loss (enforcing u(x, 0) = sin(pi * x))
    u_0_pred = net(x_0, t_0)
    initial_loss = torch.mean((u_0_pred - u_0)**2)

    return pde_loss + boundary_loss + initial_loss
```

```python
# True solution of the heat equation (analytical solution)
def true_solution(x, t, alpha):
    return torch.sin(np.pi * x) * torch.exp(-np.pi**2 * alpha * t)
```

```python
alpha = 0.1
net = PINN()
optimizer = optim.Adam(net.parameters(), lr=1e-3)
```

```python
# Collocation points in the interior (for PDE loss)
x_collocation = torch.rand((batch_size, 1), requires_grad=True)
t_collocation = torch.rand((batch_size, 1), requires_grad=True)

# Boundary condition points (x_b = 0, 1 for boundary loss)
x_b = torch.cat([torch.zeros((batch_size//2, 1)), torch.ones((batch_size//2, 1))], dim=0)
t_b = torch.rand((batch_size, 1))
u_b = torch.zeros((batch_size, 1))  # u(0, t) = u(1, t) = 0

# Initial condition points (for initial loss)
x_0 = torch.rand((batch_size, 1))
t_0 = torch.zeros((batch_size, 1))  # at t = 0
u_0 = torch.sin(np.pi * x_0)  # u(x, 0) = sin(pi * x)
```

```python
for epoch in range(epochs):
    optimizer.zero_grad()

    # Compute loss
    loss = loss_function(net, x_collocation, t_collocation, alpha, x_b, t_b, u_b, x_0, t_0, u_0)

    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f'Epoch {epoch}: Loss = {loss.item()}')
```

```python
# Choose some test points for comparison
x_test = torch.linspace(0, 1, 100).reshape(-1, 1)
t_test = torch.full((100, 1), 0.5)  # compare at t=0.5

# Predicted solution from the trained network
u_pred = net(x_test, t_test).detach().numpy()

# True solution
u_true = true_solution(x_test, t_test, alpha).detach().numpy()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test.numpy(), u_true, label='True Solution', color='blue')
plt.plot(x_test.numpy(), u_pred, label='Predicted Solution',linestyle = '--', color='red')
plt.title('True vs Predicted Solutions for Heat Equation using PINN')
plt.xlabel('x')
plt.ylabel('u(x, t)')
plt.legend()
plt.show()
```


The two methods for solving the 1D heat equation with neural networks take quite different approaches. The first method uses a traditional feedforward neural network, which focuses mainly on minimizing the error related to the heat equation's predictions. It does this by randomly sampling points without explicitly addressing the boundary and initial conditions, which can make it challenging to ensure the solution is physically accurate. On the other hand, the second method, known as Physics-Informed Neural Networks (PINNs), takes a more integrated approach. It incorporates the underlying physics directly into its training process by adding terms for the residuals of the heat equation, boundary conditions, and initial conditions to the loss function. This allows PINNs to generate solutions that naturally adhere to both the governing equations and the necessary conditions, resulting in more reliable predictions. Consequently, while the traditional approach might need a lot of training data and may struggle to meet physical constraints, PINNs tend to be more efficient and dependable, achieving accurate solutions more quickly with fewer training samples by leveraging the fundamental principles of physics.
